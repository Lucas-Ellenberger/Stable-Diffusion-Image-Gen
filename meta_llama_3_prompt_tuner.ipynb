Get Llama3 model

```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    # The quantization line
    model_kwargs={"torch_dtype": torch.bfloat16, "load_in_4bit": True}
)
```

Prompt Generation

```python
examples = """Fish are aquatic animals, they are mostly found in the sea, rivers, lakes, and oceans. Nowadays, people hunt fish for their daily bread, making business from it. There are various kinds of fish, be they small or large sizes and different shapes. Some people keep them at home as a sign of fortune.
            => 'a diagram of a fish's life cycle with the following stages', 'a fisherman sitting in a boat at sea holding a fishing rod', 'an aquatic tank containing a variety of fish inside of a home'"""
messages = [
    {"role": "system", "content": "Given a text summary, create 3 suitable educational, concise, and extremely short image prompts, example:\n" + examples},
    {"role": "user", "content": summary},
]

prompt = pipeline.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
)

terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]

outputs = pipeline(
    prompt,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)
print(outputs[0]["generated_text"][len(prompt):])
```

Parse Model Output

```python
parsed_output = outputs[0]["generated_text"][len(prompt):] # string of prompts
prompts = []
temp = parsed_output # temp array for searching through
for _ in range(3):
  start = temp.find('\'') + 1
  end = temp.find('\'', start, len(temp))
  prompts.append(temp[start:end])
  temp = temp[end + 2:]

for i in range(3):
  print(prompts[i])
```
